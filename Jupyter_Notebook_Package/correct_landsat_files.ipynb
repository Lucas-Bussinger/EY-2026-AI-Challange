```python
import pandas as pd

```

```python
train_merged = pd.read_csv("../Datasets/landsat_features_more_bands_train_merged.csv")
train_original = pd.read_csv("../Datasets/landsat_features_training.csv")
```

```python
# 1. Padronizar as datas (ESSENCIAL para que o merge funcione)
# Certifique-se de que ambos estão no mesmo formato datetime
train_original['Sample Date'] = pd.to_datetime(train_original['Sample Date'], dayfirst=True)
train_merged['Sample Date'] = pd.to_datetime(train_merged['Sample Date'], dayfirst=True)

# 2. Definir as chaves
chaves = ['Latitude', 'Longitude', 'Sample Date']

# 3. Criar o novo dataframe ordenado
# Pegamos apenas as colunas de identificação do original e trazemos os dados do merged
# Isso garante que a ordem das linhas seja IDENTICA ao original
train_merged_ordenado = pd.merge(
    train_original[chaves],  # Mantém a ordem original
    train_merged,            # Traz os dados extras
    on=chaves,               # Usa a chave tripla
    how='left'               # Garante que a ordem da esquerda (original) seja preservada
)

# 4. Verificar se o número de linhas continua igual
print(f"Linhas no original: {len(train_original)}")
print(f"Linhas no novo ordenado: {len(train_merged_ordenado)}")

# 5. Salvar o resultado
train_merged_ordenado.to_csv("../Datasets/landsat_features_more_bands_train.csv", index=False)
```
